```python
# CONCLUSION = Colorbars are set for each subplot, not overall, nor per row or column.
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='white', context='talk')

def scatter(x, y, c, **kwargs):
    plt.scatter(x, y, c=c, **kwargs)
    
tips = sns.load_dataset('tips')
g = sns.FacetGrid(tips, col='sex', row='smoker', sharex=False, size=3.5)
g = g.map(scatter, 'tip', 'total_bill', 'tip', cmap='gray_r')
```

```python
tips.loc[(tips.smoker == 'Yes') & (tips.sex == 'Male'), 'tip'] += 100
g = sns.FacetGrid(tips, col='sex', row='smoker', sharex=False, size=3.5)
g.map(scatter, 'tip', 'total_bill', 'tip', cmap='gray_r')
```
