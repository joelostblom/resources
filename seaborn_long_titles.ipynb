#Seborn long titles alter plot appearence

```python
import seaborn as sns
%matplotlib inline
sns.set_style('white')

iris = sns.load_dataset('iris')
```

##Long titles changes plot appearance

```python
#works as expected
g = sns.FacetGrid(iris, col='species', col_wrap=3)
g.map(sns.distplot, 'petal_length')

#smaller
iris['long_species'] = iris.species.str[:] * 5
g = sns.FacetGrid(iris, col='long_species', col_wrap=3)
g.map(sns.distplot, 'petal_length')

#even narrower and inverse xaxis direction
iris['long_species'] = iris.species.str[:] * 10
g = sns.FacetGrid(iris, col='long_species', col_wrap=3)
g.map(sns.distplot, 'petal_length')
```

##Matplotlib behaves the same with `plt.tight_layout()`

Without `plt.tight_layout()`, the subplots are still close together, even if the entire figure is shrinking and the x-axes never change direction.

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1,3, figsize=(8,4))
for ax in axes.ravel():
    ax.hist(iris.petal_length)
    ax.set(title='long_title' * 5 )
    sns.despine()
plt.tight_layout()
    
fig, axes = plt.subplots(1,3, figsize=(8,4))
for ax in axes.ravel():
    ax.hist(iris.petal_length)
    ax.set(title='long_title' * 10 )
    sns.despine()
plt.tight_layout()
    
fig, axes = plt.subplots(1,3, figsize=(8,4))
for ax in axes.ravel():
    ax.hist(iris.petal_length)
    ax.set(title='long_title' * 20 )
    sns.despine()
# plt.tight_layout()

fig, axes = plt.subplots(1,3, figsize=(8,4))
for ax in axes.ravel():
    ax.hist(iris.petal_length)
    ax.set(title='long_title' * 50 )
    sns.despine()
# plt.tight_layout()
```
