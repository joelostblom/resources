```python
import seaborn as sns
%matplotlib inline
sns.set_style('white')
iris = sns.load_dataset('iris')

g = sns.FacetGrid(data=iris, hue='species', size = 3)
g.map(sns.distplot, 'petal_length', hist_kws={'histtype':'stepfilled'})

g = sns.FacetGrid(data=iris, hue='species', size = 3)
g.map(sns.distplot, 'petal_length', hist_kws={'linewidth':0})

g = sns.FacetGrid(data=iris, hue='species', size = 3)
g.map(sns.distplot, 'petal_length', kde=False, hist_kws={'histtype':'stepfilled'})

g = sns.FacetGrid(data=iris, hue='species', size = 3)
g.map(sns.distplot, 'petal_length', kde=False, norm_hist=True, hist_kws={'histtype':'stepfilled'})

g = sns.FacetGrid(data=iris, hue='species', size = 3)
g.map(sns.distplot, 'petal_length', kde=False, norm_hist=True, hist_kws={'linewidth':0})
g.axes.ravel()[-1].legend()
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('../phd/ce/data/KinaseScreen.csv')
df = df.loc[np.random.choice(10000, 1000)]
```

```python
g = sns.FacetGrid(df, hue='Condition', sharey=False, size=4, xlim=(0,1000))
g.fig.suptitle('AvgIntenCh3', size=14, y=1.02)
g.map(sns.distplot, 'AvgIntenCh3', hist_kws={'histtype':'stepfilled'})
g.add_legend()
```

```python
import matplotlib.pyplot as plt
import seaborn.apionly as sns
iris = sns.load_dataset('iris')

plt.figure()
for species, species_df in iris.groupby("species"):
    plt.hist(species_df["petal_length"].values, normed=True, histtype="stepfilled")
plt.legend(sorted(iris.species.unique()))

plt.figure()
for species, species_df in iris.groupby("species"):
    plt.hist(species_df["petal_length"].values, histtype="stepfilled")
    
plt.figure()
for species, species_df in iris.groupby("species"):
    plt.hist(species_df["petal_length"].values, normed=True)
    
plt.figure()
iris.loc[iris.species == 'setosa', 'species'] = 'wsetosa'
for species, species_df in iris.groupby("species"):
    plt.hist(species_df["petal_length"].values, normed=True, histtype="stepfilled")
plt.legend(sorted(iris.species.unique()))
```
