# From spreadsheets to Pandas

## Conceptual understanding

Spreadsheet programs, such as MS Excel and Libre/OpenOffice, have their funcionality sectioned into a menu/ribbon driven interface. In programming languages, such as R and Python, all the functionality is accessed by typing the name of functions directly instead of finding the functions in menus, which might initially seem intimidating and non-intuitive for people who are used to the menu-driven approach. This is a brief tutorial for people with a background in spreadhsheet software, that are interestesting in learning basic R or Python.

Briefly, spreadheets are great for viewing and entering small datasets and creating simple visualizations fast. However, it can be tricky to work with spreadsheets to create publication ready figures, automate analysis worflows, perform advanced calulcations, and clean datasets robustly. So even if your data will be in the form of a spreadsheet, it can be beneficial to pick up som basic programming skills to help with the analyses of that data.

## Programming basics

Just like in spreadsheet software, the basic installation of R and Python includes fundamental math operations, e.g. adding numbers together.

```python
4 + 4 
```

If you want to access extra functionality from a spreadhseet program, you need to click the menus/ribbons and select the tool you want to use. All charts are in one menu, text layout tools in another and data analyses tools in a third. Programming languages such as R and Python have so many tools and functions so that they would not fit in a menu. So instead of clicking `Insert >> Image` and then chose the file, you will type something similar to `insert.image('<filename>')`.

Since there are so many tools and functions available in Python and R, it is also unnecessary to install and load all of them at once. Instead, more advanced functionality is grouped into separate packages, which can be accessed by by typing `import <package_name>` in Python or `library('<package_name>')` in R. You can think of this as that you are telling the program which menu items you want to use (similar to how Excel hides the `Developer menu` by default since most people rarely use it and you need activate it in the settings if you want to access its functinoaliy).

For convenience, I will focus mostly on Python for the rest of this tutorial. The functionality described here is available in R as well, but sometimes under a different library or function name.

Just like in spreadhsheet software menus, there are lots of different tools within each Python package. For example, if I want to use numerical python functions, I can import the numerical python module, `numpy` for short. I can then access all the functions by writing `numpy.<function_name>`.

```python
import numpy

numpy.mean([1,2,3,4,5])
```

Once you start out using Python, you don't know what functions are availble within each package. Luckily, if you type `numpy.`<kbd>tab</kbd> (that's 'numpy' + period-key + tab-key) a small menu will popup that shows you all the available funcitons in that module. As I mentioned earlier, there are plenty of available functions and it can be helpful to filter the menu by typing the beginning letters of the function.

I took a screenshot of how this tab-completion looks, and below I use the `Image` function from the `IPython.display` package to display this screenshot.

```python
from IPython.display import Image
Image('./tab-completion.png')
```

To get more info on the function you want to use, you can type out the full name and then press <kbd>Shift + Tab</kbd> once to bring up a help dialogue and again to expand that dialogue. For `numpy.mean` below, we can see that the first argument that needs to be inputted to the function is `a`, which should be 'array-like'. An array is essentially just a sequence of numbers. In our case above, we passed a sequenced of number in brackets `[]`, which in Python means that these numbers are in a list, which is a sequence of number just like an array.

We could also have defined the `a`-array as a tuple: `(1,2,3,4,5)`, or the specific numpy array format: `numpy.array(1,2,3,4,5)`. It is not important to know all the differences between these formats for now, but in short numpy arrays has the most options for numerical arrays (such as in built mean functions: `a.mean()`) and lists are most commonly used for all other types of arrays.

```python
Image('./shift-tab-help.png')
```

If you need a more extensive help dialog, you can click  <kbd>Shift + Tab</kbd> four time or just type `?numpy.mean`.

## Data tables/frames
The python package that is most commonly used to work with spreadsheets is called `pandas`, the name is derived from "panel data", an econometrics term for multidimensional structured data sets ^1 (ref wiki). Spreadsheets are easily loaded into pandas from `.csv` or Excel formats. I don't have any good data set lying around, so I am loading a public online sample dataset. I store my dataset in a variable called `iris`, that I can use for downstream manipulations.

```python
import pandas as pd

iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

To have a quick peak at the data, I can type `iris.head(<number_of_rows>)`. Notice that I do not have to use the `pd.`-syntax for this. `iris` is now a pandas dataframe, and all pandas data frames have a number of inbuilt functions that can be used directly from the dataframe instead of by calling `pandas` separately.

```python
iris.head(10)
```

I can find out how many rows and columns my data set has

```python
iris.shape
```

If you want to be able to see all the rows at a time, you can either display them all in this notebook, or just have your excel spreadsheet open next to this notebook. You can easily get a basic description of the data in the different columns.

```python
iris.describe()
```

Instead of looking at the general description of the data, we might want to make more specific comparisons, such as between the mean values between the different species.

```python
# we will reuse this data frame, so I save it as a variable 
species_comparison = iris.groupby('species').mean()
species_comparison
```

We can see that there are clear differences, but they might be even clearer if we display them graphically in a chart.

## Plotting

```python
%matplotlib inline # To display the plot in the notebook instead of in a separate window

species_comparison.plot(kind='bar')
```

Depending in what your are interesting in showing, it could be useful to have the species as the different colors and the columns along the xaxis. We can easily achieve this by transposing (`.T`) our data frame.

```python
species_comparison.T.plot(kind='bar')
```

The graph is a bit ugly. Instead of changing the appearance of each bar and line one by one, we can easily apply a new style to entire graph. To change the style, we need to use the specific plotting package `matplotlib`, which is also what pandas uses for its plots.

```python
import matplotlib.pyplot as plt

# the with-keyword is used to apply this style only to the indented code instead of all plots in the notebook
with plt.style.context('ggplot'):
    species_comparison.T.plot(kind='bar')
```

That looks better, but still not fantastic. A more powerful plotting library is `seaborn`. Like `pandas`, `seaborn` builds upon `matplotlib`, and adds new styles, powerful new plot types and also some statistical measures. As you will notice, we do not need to call `groupby` or `mean` on the `iris`-dataframe when plotting with `seaborn`. Instead we control these options from `seaborn` with the plot type we chose (barplot = mean automatically) and the `hue`-parameter is analogous to `groupby`.

```python
import seaborn as sns

sns.barplot(y='sepal_length', x='species', data=iris)
```

To get the exact same plot as we had previously, we need to reshape our data frame into the [tidy data - format](http://vita.had.co.nz/papers/tidy-data.pdf). This is suggested for ease of use.... INCLUDE A FIGURE

Luckily, we can use pandas inbuilt function to make the data into the tidy-format.

```python
iris_melt = pd.melt(iris, id_vars='species')
iris_melt.head()
```

Now we can recreate the plot we made earlier.

```python
sns.barplot(x='variable', y='value', hue='species', data=iris_melt)
```

Again, it is easy to change the deafult style.

```python
sns.set(style='white', context='notebook')
sns.barplot(x='variable', y='value', hue='species', data=iris_melt)
```

 Now that we know how to use seaborn, we can create more insigthful plots with less effort.

```python
sns.swarmplot(x='variable', y='value', hue='species', data=iris_melt, split=True, palette='Set2')
sns.despine() # removes the right and top black border
```

```python
with sns.plotting_context('talk') and sns.axes_style("ticks"):
    sns.boxplot(y='variable', x='value', hue='species', data=iris_melt, color='c', )
    sns.stripplot(y='variable', x='value', hue='species', data=iris_melt, size=2.5, palette=['k']*3, jitter=True, split=True)
sns.despine(trim=True)
```

It is also possible to get too fancy and accidentally hide important messages in the data. However, the fact that you now have access to several ways to plot your data forces you to consider what is actually important and how you can best communicate that message, rather than always making the same plot and not thinking about the strengths and pitfalls of that plot type.
